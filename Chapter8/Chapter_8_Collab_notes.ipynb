```python
ratings = pd.read_csv(path/'u.data', delimiter='\t', header=None,
                      names=['user','movie','rating','timestamp'])
ratings.head()
movies = pd.read_csv(path/'u.item',  delimiter='|', encoding='latin-1',
                     usecols=(0,1), names=('movie','title'), header=None)
ratings = ratings.merge(movies)
ratings.reset_index(drop=True)
```

1. What problem does collaborative filtering solve?  
Collaborative filtering tries to to solve the problem of guessing ratings/valuations from users for unknown items (movies, restaurants etc.). 
2. How does it solve it?  
Each user and each item is assigned some latent factors that are calibrated based on the values of known ratings for that user and known ratings for that movie. Then based on these factors the "match" between the user and new items can be calculated.
3. Why might a collaborative filtering predictive model fail to be a very useful recommendation system?  
Because of lack of information. If the system does not have enough information about a user it cannot recommend well.   
Because of overfitting to a set of users that are always scoring their favourite films/items highly which attracts same users that continue the high scoring which perpetuates the issue.  
Not enough latent factors to accurately describe the users/movies.
